Python Basic

Q1- What is Python, and why is it popular

Python is a high-level, interpreted programming language known for its simple syntax, readability, and wide range of libraries.
    
Popular because:
1. Easy to learn and write
2. Versatile
3. Strong community support

Q2 - What is an interpreter in Python

An interpreter reads and executes Python code line-by-line, converting it into machine code.

Q3 - What are pre-defined keywords in Python

Pre defined keywords are reserved words that have special meaning in Python. Example if, else, while, etc

Q4 - Can keywords be used as variable names

No, keywords cannot be used as variable names because they have predefined meanings in the language.

Q5 - What is mutability in Python

Mutability means whether an object content can be changed after it is created.

Q6 - Why are lists mutable, but tuples are immutable

Lists are designed to be dynamic and changeable, useful for situations where data updates.
Tuples are fixed and secure, ideal for constant data or keys in dictionaries.

Q7 - What is the difference between “==” and “is” operators in Python

== checks value equality and is checks identity

Q8 - What are logical operators in Python

Logical operators are used to combine conditional statements:

1. and - True if both conditions are true
2. or - True if at least one condition is true

Q9 - What is type casting in Python

Type casting means converting a value from one data type to another.

Q10 - What is the difference between implicit and explicit type casting?

Implicit - Done by python itself
Explicit - Manually by the programmer

Q11 - What is the purpose of conditional statements in Python

Conditional statements allow the program to make decisions and execute code based on conditions

Q12 - How does the elif statement work?

elif stands for “else if”. It checks another condition if the previous if was false.

Q13 - What is the difference between for and while loops?

1. for loop: used when you know in advance how many times to loop (e.g., for items in a list)
2. while loop: used when you loop until a condition becomes false (condition-based)

Q14 - Describe a scenario where a while loop is more suitable than a for loop

When you don’t know beforehand how many times the loop will run.
1. Example: Keep asking for a password until the correct one is entered.
