Q4 - Write a program that checks if a given word is a Python keyword.

```python
import keyword

# Take input from the user
word = input("Enter a word: ")

# Check if it is a Python keyword
if keyword.iskeyword(word):
    print(f"'{word}' is a Python keyword.")
else:
    print(f"'{word}' is NOT a Python keyword.")
```

Q10 - Write code to demonstrate type casting with list elements

```python
str_list = ["1", "2", "3", "4", "5"]

# Convert strings to integers
int_list = [int(x) for x in str_list]

# Convert integers to floats
float_list = [float(x) for x in int_list]

print("Original list (strings):", str_list)
print("Converted to integers:", int_list)
print("Converted to floats:", float_list)
```

Q13 - Write a Python program to find the sum of all even numbers between 1 and
50

```python
total = 0
for number in range(1, 51):
    if number % 2 == 0:
        total += number

print("The sum of all even numbers between 1 and 50 is:", total)
```

Q14 - Write a program to reverse a string using a while loop.

```python
# Get input from user
text = input("Enter a string: ")

reversed_text = ""
index = len(text) - 1
while index >= 0:
    reversed_text += text[index]
    index -= 1
print("Reversed string:", reversed_text)
```

Q 15 -  Write a Python program to calculate the factorial of a number provided by the user using a while loop.

```python
num = int(input("Enter a number: "))
factorial = 1
i = 1
while i <= num:
    factorial *= i
    i += 1
print("Factorial of", num, "is:", factorial)
```
